# Tarefa 1

### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
sinasc_1 = sinasc_raw[['munResNome','IDADEMAE','IDADEPAI']] #criando o data frame com a idade dos pais e o municipio
sinasc_1
```

```python
sinasc = sinasc_1.groupby('munResNome') #atribui a distribuiçao por municipios a variabel sinasc
sinasc.mean() #calcula a media das colunas IDADEMAE e IDADEPAI por municipio.
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
#separando o dataframe com as variaveis SEXO,PESO, ESCMAE e DTNASC.
sinasc_2 = sinasc_raw[['SEXO', 'ESCMAE','DTNASC','PESO']]
sinasc_2
```

```python
niver = sinasc_2[sinasc_2['DTNASC']=='2019-01-31'] #filtra pela minha data de nascimento 31/01
```

```python
peso_med = niver.groupby(['ESCMAE','SEXO'])['PESO'].mean().unstack() #usa o filtro anterior e agrupa por escolaridade e sexo, calc a media dos pesos
peso_med
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
#separando o dataframe com as variaveis a serem utilizadas: IDADEMAE, IDADEPAI,munResNome.
sinasc_3 = sinasc_raw[[ 'munResNome','IDADEMAE', 'IDADEPAI']]
sinasc_3
```

```python
#municipio que menos nasceu bebes
sinasc_3['munResNome'].value_counts().sort_values()
```

```python
# O municipio com menos nascimento foi "Município ignorado - RO"
# acredito ser apenas uma informaçao generica, entao o municipio com menor quantidade de nacismentos é "Castanheiras".
```

```python
#agrupando os dados por municipio e agragando as idades max,med e min das maes e pais
sinasc_3.dropna().groupby('munResNome').agg([('mínimo','min'),('média','mean'),('máximo', 'max')])
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
#df
sinasc_4 = sinasc_raw[['munResNome','DTNASC','QTDFILVIVO','IDADEMAE','IDADEPAI']].dropna()
sinasc_4
```

```python
#municipios que nasceram mais bebes no mes de março

bebes_mar = sinasc_4['DTNASC'] = pd.to_datetime(sinasc_4['DTNASC'])
bebes_mar = bebes_mar.dt.month == 3
bb_mar = sinasc_4[bebes_mar]
```

```python

mun_mais_nasc = bb_mar['munResNome'].value_counts().idxmax()
mun_mais_nasc
```

```python
count_bb = bb_mar['munResNome'].value_counts().sort_values(ascending = False)
count_bb
```

```python
print('O municipio com maior quantidade de nascimentos no mes de março foi:', mun_mais_nasc,'com',count_bb.iloc[0], 'nascimentos.')
```

```python
bb_mar.dtypes
```

```python
#calculando a idade media dos pais e a quantidade de filhos vivos nesse municipio
bb_mar[bb_mar['munResNome'] == 'Cacoal'].groupby(['munResNome', 'DTNASC']).agg([('Minimo',min),('Media','mean'),('Maximo',max)])
```
