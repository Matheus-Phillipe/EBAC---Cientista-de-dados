````python
#importanto as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
#lendo o arquivo csv previsao de renda
df = pd.read_csv('/kaggle/input/previsao-de-renda/previsao_de_renda.csv')

df
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
#verificando a distribuiçao das variaveis ao longo do tempo
#começando pela variavel sexo

ax = sns.countplot(x='data_ref',  hue = 'sexo', data=df)
plt.xticks(rotation=90)
````

Comparando a variação da variável sexo ao longo do tempo, é possível notar que o padrão se manteve, em sua maioria com aproximadamente 650 pessoas do sexo feminino para 350 pessoas do sexo masculino. Então essa variável se manteve estavel ao longo do tempo.

````python
#variavel educaçao
ax = sns.countplot(x='data_ref',  hue = 'educacao', data=df)
plt.xticks(rotation=90)
````

Ao longo do tempo é possível perceber que as pessoas com ensino superior completo foi aumentando em relação ao início. Inicialmente a maioria das pessoas da base de dados tinham apenas o secundário completo, porém, a partir da data 01/09/2015, pessoas com ensino superior completo cresceram e praticamente se igualaram aos com secundário. Considerando que para cada mês foram inseridas 1000 pessoas.

````python
#variavel estado civil

ax = sns.countplot(x='data_ref',  hue = 'estado_civil', data=df)
plt.xticks(rotation=90)
````

Para a variável estado civil, não ocorreu uma variação significativa, sendo em sua maioria pessoas casadas.

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

As variáveis analisadas foram sexo, educação e estado civil em relação a renda ao longo do tempo. 

````python
#verificando os perfis medios no tempo para a variavel sexo em relaçao a renda

ax = sns.pointplot(x="data_ref", y="renda", hue = 'sexo',
                   data=df, dodge=True, errorbar=('ci',95))
plt.xticks(rotation=90)
````

Primeiro olhando apenas para a variável ‘sexo’: diferente do gráfico das contagens onde a maioria inserida na base são mulheres, ao comprar diretamente com a renda, vemos que a renda masculina se mante superior as mulheres. Mesmo com essa inversão de maioridade, se manteve bem estável ao longo do tempo, sendo uma boa variável para uma modelagem.

````python
#verificando os perfis medios no tempo para a variavel educacao em relacao a renda

ax = sns.pointplot(x="data_ref", y="renda", hue = 'educacao',
                   data=df, dodge=True, errorbar=('ci',95))
plt.xticks(rotation=90)
````

````python
#separando as pessoas com pos graduacao para remove-las
pos_grad = df[df['educacao']=='Pós graduação']
pos_grad
````

````python
#separando os indices das pessoas com pós para remover
indices = pos_grad.index 
df1 = df.drop(indices)
````

````python
#novo DataFrame sem as pessoas com pós
df1['educacao'].value_counts()
````

````python
#refazendo a analise com o novo df
ax = sns.pointplot(x="data_ref", y="renda", hue = 'educacao',
                   data=df1, dodge=True, errorbar=('ci',95))
plt.xticks(rotation=90)
````

Já em relação a educação as pessoas que estão em sua minoria com pós-graduação completa, são as que tiveram a maior renda ao longo do tempo e essa analise se mostrou muito instável. Como as pessoas que possuem pós-graduação completa são uma porcentagem muito baixa, refiz a analise com um novo dataframe sem elas. No novo dataframe é possível usar para uma modelagem uma vez que a variação ao longo do tempo não foi significativa. Porem é passivo de analisar o modelo ao longo do tempo para confirmar a eficiencia do modelo.

````python
#verificando os perfis medios no tempo para a variavel estado civil em relacao a renda

ax = sns.pointplot(x="data_ref", y="renda", hue = 'estado_civil',
                   data=df, dodge=True, errorbar=('ci',95))
plt.xticks(rotation=90)
````

O estado civil das pessoas foi a variável mais instável ao longo do tempo em relação a renda, não sendo eficiente para uma modelagem.
