```python
# importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/dados-sinasc/dados_SINASC.csv') # lendo o dataset
```

```python
df # exibe o dataset
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(17,10)) #define o tamanho da figura 
fig , axes = plt.subplots(2,3) # cria uma matris 2x3 

sns.countplot(ax = axes[0,0], x = 'APGAR1', data = df) # grafico na posiçao [0,0] para a variavel APGAR1
sns.countplot(ax = axes[0,1], x = 'APGAR5', data = df)# grafico na posiçao [0,1] para a variavel APGAR5
sns.countplot(ax = axes[0,2], x = 'SEXO', data = df) # grafico na posiçao [0,2] para a variavel SEXO
gest = sns.countplot(ax = axes[1,0], x = 'GESTACAO', data = df) # grafico na posiçao [1,0] para a variavel GESTACAO
sns.countplot(ax = axes[1,1], x = 'ESTCIVMAE', data = df) # grafico na posiçao [1,1] para a variavel ESTCIVMAE

gest.set_xticklabels(gest.get_xticklabels(), rotation=90, horizontalalignment='right') #rotaciona o eixo x da variavel GESTACAO em 90 graus
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all') #fecha os outros graficos
plt.rc('figure', figsize=(10,4)) #novo tamanho para figura
sns.barplot(x = 'APGAR5', y = 'CONSULTAS', data = df) # exibe o grafico de barras com apgar no eixo X e consultas no eixo Y
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all') # fecha todos os graficos
plt.rc('figure', figsize=(13,10)) # novo tamanho para a figura

fig, axes = plt.subplots(2,2) # cria uma matriz [2,2]

sns.countplot(ax = axes[0,0], x = 'SEXO', data = df) # grafico na posiçao [0,0] para a variavel SEXO
gest = sns.countplot(ax = axes[1,0], x = 'GESTACAO', data = df) # grafico na posiçao [1,0] para a variavel GESTACAO
sns.countplot(ax = axes[0,1], x = 'ESTCIVMAE', data = df) # grafico na posiçao [0,1] para a variavel ESTCIVMAE
sns.countplot(ax = axes[1,1], x = 'CONSULTAS', data = df) # grafico na posiçao [1,1] para a variavel CONSULTAS

gest.set_xticklabels(gest.get_xticklabels(), rotation=90, horizontalalignment='right') #rotaciona o eixo x da variavel GESTACAO em 90 graus
```
